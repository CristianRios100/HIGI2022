```python
import numpy as np
import math as mat
import matplotlib.pyplot as plt
import sympy
```

```python
def f_planet(x0, vx0, y0, z0, t0=0, dt=0.01 ,steps=100):
    G = 4*(mat.pi**2)
    M = 1
    rp = mat.sqrt( x0**2 + y0**2 + z0**2 ) 
    x = [x0] 
    t = [t0]
    for i in range(1, steps):
        Vxi = vx0 - ( (G*M) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        x.append(xi)
        t.append(t0+dt*i)
        x0 = xi
        vx0 = Vxi 
    return x, t
```

```python
#Mercurio
x, t = f_planet(0.0508011, 0.022318, -0.4011642, -0.2195576, steps=100
               )
plt.plot(t , x)
```
